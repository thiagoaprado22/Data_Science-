```python
import pandas as pd
url1 = "https://ckan.pbh.gov.br/dataset/05f673cf-aa81-4e03-bb3d-7c1ea91e7bec/resource/bd8893c9-ee50-49ac-8106-cab675180cf5/download/enderecos_nordeste.csv"
url2 = "https://ckan.pbh.gov.br/dataset/e0aa516b-8da9-405a-812d-17ced7d18baa/resource/3e1939f7-5786-4ce6-8795-11f04e4fa837/download/enderecos_barreiro.csv"
url3 = "https://ckan.pbh.gov.br/dataset/9ba40bcf-dfe6-440a-9b43-e4fda65d3248/resource/1502ba75-4d4d-4aa6-9cf9-d3802d8008ca/download/enderecos_leste.csv"
url4 = "https://ckan.pbh.gov.br/dataset/4717ec09-52dd-480e-8e46-cb356b8c0c22/resource/688ff522-482d-499b-ac59-32479283cde3/download/enderecos_norte.csv"
url5 = "https://ckan.pbh.gov.br/dataset/34b5a77f-3ac2-440d-9727-3b9407f3cf2e/resource/ab05c850-5fb0-4b37-bd60-8bf93d35978b/download/enderecos_oeste.csv"
url6 = "https://ckan.pbh.gov.br/dataset/10ea81f7-a4e3-43e3-b95b-72b1eb8ff85a/resource/3d0fb1e1-b0b7-4378-95a3-f37193566969/download/enderecos_pampulha.csv"
url7 = "https://ckan.pbh.gov.br/dataset/de7088d4-cb67-45d7-96ce-f24f99705ff5/resource/f84d29a1-239b-491a-a9a7-68bdd382de78/download/enderecos_noroeste.csv"
dados_nordeste = pd.read_csv(url1)
dados_barreiro = pd.read_csv(url2)
dados_leste = pd.read_csv(url3)
dados_norte = pd.read_csv(url4)
dados_oeste = pd.read_csv(url5)
dados_pampulha = pd.read_csv(url6)
dados_noroeste = pd.read_csv(url7)
```

```python
## Importanto a segunda data frame
```

## Data Frame 1 - Regiões de BH - PBH

```python
bigdata = [dados_nordeste , dados_barreiro, dados_leste, dados_norte, dados_oeste, dados_pampulha, dados_noroeste]
dados_bh = pd.concat(bigdata)
dados_bh
```

```python
dados_bh.info()
```

```python
dados_bh.drop(columns= ['SIGLA_TIPO_LOGRADOURO', 'NUMERO_IMOVEL', 'LETRA_IMOVEL' ,'CEP', 'GEOMETRIA'])
```

```python
dados_bh["NOME_LOGRADOURO"] = dados_bh.NOME_LOGRADOURO.astype(float)
```

```python
dados_bh["NOME_LOGRADOURO"].unique()
```

```python
dados_bh_bairro = dados_bh.groupby("NOME_BAIRRO_POPULAR").mean()
dados_bh_bairro
```
